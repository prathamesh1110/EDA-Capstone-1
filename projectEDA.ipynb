**Q1.What is the annual number and percentage of arriving visitors?**

```python
df['arrival_date_year'].value_counts()
```

```python
yearly_arrival_of_guests = df.groupby(['arrival_date_year','hotel'])['hotel'].count().unstack()
yearly_arrival_of_guests.plot(kind='bar',color = ['green','red'],figsize=(10,8),fontsize = 12)
plt.title('Annual numbers of guest arrived')
plt.xlabel('years')
plt.ylabel('no. of guest')
plt.show()

percent = df["arrival_date_year"].value_counts()
percent.plot.pie(fontsize= 18,autopct="%.1f%%",figsize=(10,8),textprops={'fontsize': 20})
plt.title('Year comparison')
plt.ylabel('% of Bookings')
plt.xlabel('Year')
```

**Conclusion**- 
**1)** 2016 has more number of visitors than 2015 and 2017.

**2)** After 2015 number of visitors increased in 2016 but again  in 2017 it is decreased.  

**Q2.Which marketing Sector should be targeted for advertisement??**

```python
df['market_segment'].value_counts()
```

```python
df['distribution_channel'].value_counts()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data = df, x = 'market_segment').set_title('Types of market segment', fontsize = 20)
plt.show()
sns.countplot(data = df, x = 'distribution_channel').set_title('Types of distribution_channel', fontsize = 20)
```

**Conclusion**-

1.Frow above we see that the majority of the distribution channels and market segments involve travel agencies (online or offline).

2.We have to focus on targeting these travel agencies website and work with them since majority of the visitors tend to reach out to them.



**Q3-Which hotel has more cancelled bookings?**

```python
df['is_canceled'].value_counts()
```

```python
(df["is_canceled"].value_counts()).plot.pie(autopct="%.1f%%");plt.legend()
plt.title("Booking cancelled")
plt.xlabel("canceled and Not canceled")
plt.legend()
plt.show()
```

**Q3.Which month has highest and lowest month bookings?**

```python
df['arrival_date_month'].value_counts()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=df['arrival_date_month']).set_title('Number of monthly bookings')
plt.show()
plt.figure(figsize=(15,8))
sns.countplot(x='arrival_date_month',hue='hotel',data=df).set_title('Number of monthly bookings')

```

**Conclusion**-
1. August month has highest number of bookings.
2. January month has lowest number of bookings.
