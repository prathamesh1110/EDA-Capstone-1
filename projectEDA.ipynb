**Q.What is the annual number and percentage of arriving visitors?**

```python
df['arrival_date_year'].value_counts()
```

```python
yearly_arrival_of_guests = df.groupby(['arrival_date_year','hotel'])['hotel'].count().unstack()
yearly_arrival_of_guests.plot(kind='bar',color = ['green','red'],figsize=(10,8),fontsize = 12)
plt.title('Annual numbers of guest arrived')
plt.xlabel('years')
plt.ylabel('no. of guest')
plt.show()

percent = df["arrival_date_year"].value_counts()
percent.plot.pie(fontsize= 18,autopct="%.1f%%",figsize=(10,8),textprops={'fontsize': 20})
plt.title('Year comparison')
plt.ylabel('% of Bookings')
plt.xlabel('Year')
```

**Conclusion**- 
**1)** 2016 has more number of visitors than 2015 and 2017.
**2)** After 2015 number of visitors increased in 2016 but again  in 2017 it is decreased.
